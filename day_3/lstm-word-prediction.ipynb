```python
text = """Hello, I'm Kush Doshi, a passionate Python Developer and tech community leader. With a track record of founding Google Developer Student Clubs, organizing impactful events, and contributing to prominent conferences, I bring a wealth of experience.

I excel in crafting practical solutions, backed by internships at renowned firms like Madvise Infotech and Brainy Beam Technologies. Academic accolades, including the Dewang Mehta IT Award, underscore my dedication to excellence. Driven by a vision to harness technology for positive change, I'm committed to continual growth and making a lasting impact in the tech industry."""
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()

tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
input_sequences = []
for sentence in text.split("\n") :
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(tokenized_sentence)) :
        input_sequences.append(tokenized_sentence[: i+1])
```

```python
input_sequences
```

```python
max_length = max([len(x) for x in input_sequences])
```

```python
max_length
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_input_sequences = pad_sequences(input_sequences, maxlen = max_length, padding = 'pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:, :-1]
y = padded_input_sequences[:, -1]
```

```python
from tensorflow.keras.utils import to_categorical

y = to_categorical(y, num_classes = len(tokenizer.word_index)+1)
```

```python
X.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Input
```

```python
model = Sequential()
model.add(Input(shape = (56,)))
model.add(Embedding(75, 100))
model.add(LSTM(150))
model.add(Dense(75, activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = ['accuracy'])
```

```python
model.fit(X, y, epochs = 100)
```

```python
test = "passionate"
for i in range(10) :

    token_text = tokenizer.texts_to_sequences([test])[0]
    
    padded_token_text = pad_sequences([token_text], maxlen = 56, padding = 'pre')
    pos = np.argmax(model.predict(padded_token_text))

    for word, index in tokenizer.word_index.items() :
        if index == pos :
            test = test + " " + word
            print(test)
```
